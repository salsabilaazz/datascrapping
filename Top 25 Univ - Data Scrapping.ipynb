```python
import pandas as pd
from bs4 import BeautifulSoup
import chromedriver_autoinstaller
from selenium import webdriver
```

```python
chromedriver_autoinstaller.install()

browser = webdriver.Chrome()
```

```python
browser.get('https://www.timeshighereducation.com/world-university-rankings/2023/world-ranking/')
```

```python
soup = BeautifulSoup(browser.page_source, 'lxml')
```

```python
rows = soup.find_all('tr', class_='js-row')
```

```python
rows
```

```python
data = []
for row in rows:
    
    rank = row.find('td', class_='rank').text
    univ = row.find('a', class_='ranking-institution-title').text.strip()
    country = row.find('div', class_='location').text
    fte = row.find('td', class_='stats_number_students').text
    students_staff = row.find('td', class_='stats_student_staff_ratio').text
    intr_students = row.find('td', class_='stats_pc_intl_students').text
    fem_male = row.find('td', class_='stats_female_male_ratio').text
    
    rich = {
        'Rank' : rank,
        'University' : univ,
        'Country' : country,
        'No. of FTE Students' : fte,
        'No. of Students per Staff' : students_staff,
        'International Students' : intr_students,
        'Female : Male Ratio' : fem_male
    }
    data.append(rich)
```

```python
pd.DataFrame(data)
```

```python
pd.DataFrame(data).to_excel('Top 25 Universities in The World by THE.xlsx')
```
